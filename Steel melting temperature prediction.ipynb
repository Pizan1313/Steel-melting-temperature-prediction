# Steel melting temperature prediction

## Problem condition

In order to optimize production costs, the smelter decided to reduce electricity consumption during the steel processing stage. You have to build a model that predicts the temperature of the steel.




## Description of the processing step

Steel is processed in a metal ladle with a capacity of about 100 tons. For the ladle to withstand high temperatures, it is lined with refractory bricks from the inside. Molten steel is poured into a ladle and heated to the desired temperature with graphite electrodes. They are installed in the lid of the bucket.

Sulfur is removed from the alloy (desulfurization), the chemical composition is corrected by adding impurities, and samples are taken. Steel is alloyed - its composition is changed - by feeding pieces of alloy from a bulk materials bin or wire through a special tribe apparatus.

Before introducing alloying additives, the temperature of the steel is measured, and its chemical analysis is carried out. Then the temperature is raised for several minutes, alloying materials are added, and the alloy is purged with an inert gas. Then it is stirred and measured again. This cycle is repeated until the target chemical composition and optimum melting temperature are reached.

Then the molten steel is sent to finish the metal or enters the continuous casting machine. From there, the finished product comes out as slab blanks.

The <code>key</code> column contains the batch number in all files. Several lines in files with the same <code>key</code> value can correspond to different processing iterations.

<b>Active power</b> characterizes the rate of irreversible conversion of electrical energy into other forms of energy. In this case, thermal and electromagnetic. That is, it directly speaks about the amount of heating during steel alloying. This power can be exclusively positive. It is usually calculated by dividing the work by the time or as the product of the current square, the ohmic resistance, and the cosine of the phase angle between the current and the voltage in the network. <b>Reactive power</b> - a value that characterizes the loads created in electrical devices by fluctuations in the energy of an electromagnetic field in a sinusoidal alternating current circuit. Reactive power considers the sine of the phase angle between current and voltage. Accordingly, it can take negative values. When the reactive power of the device is positive (active-inductive character), it is considered that it consumes it, and when it is negative (active-capacitive character), it produces it. These are losses in the network that are not aimed at increasing efficiency.

## Preprocessing

<div style="border-radius: 15px; box-shadow: 2px 2px 2px; border: 1px solid green; padding: 20px">
<b>We calculate the sums of active capacities</b>
    
We do not know if the df_arc table contains power or work data (whether they have time dependencies or were measured at a point in time). Power data can be calculated either from the work A=P/t taking into account time or directly P=R*I^2 without considering the electric current's duration.
    
Therefore, we will add to the final dataset the indicator of the duration of the supply of electric current to the electrodes. If the power and time have a high correlation close to one, we will not consider it when training the model.
</div>

```python
df_arc.columns = ['key', 'start_time','end_time', 'active_power', 'reactive_power']
df_arc['duration'] = df_arc['end_time'] - df_arc['start_time']
for i in range(len(df_arc['duration'])):
    df_arc['duration'][i] = df_arc['duration'][i].seconds
df_arc_sum = pd.pivot_table(df_arc,
                             values=['active_power', 'duration'],
                             index='key',
                             aggfunc={'active_power': np.sum,
                                     'duration': np.sum})
df_arc_sum.columns = ['sum_active_power','sum_duration']
df_arc_sum.head()
```

<div style="border-radius: 15px; box-shadow: 2px 2px 2px; border: 1px solid green; padding: 20px">
<b>Let's deal with the temperature</b>
    
We will write out the batch numbers for which the data in the temperature measurement table do not meet the condition: there should be more than one measurement, one more than there were heatings, and the last measurement occurred after heating. These batch numbers will be removed from the dataset.
</div>

```python
df_temp['Время замера']= pd.to_datetime(df_temp['Время замера'])
for column in df_bulk_time.columns:
    if column[0] == 'B':
        df_bulk_time[column]= pd.to_datetime(df_bulk_time[column])
for column in df_wire_time.columns:
    if column[0] == 'W':
        df_wire_time[column]= pd.to_datetime(df_wire_time[column])
```

```python
drop_keys = []
for key in list(df_temp['key'].unique()):
    if ((df_temp[df_temp['key'] == key]['Время замера'].max() < df_arc[df_arc['key'] == key]['end_time'].max()) or
           (df_temp[df_temp['key'] == key]['Время замера'].max() == df_temp[df_temp['key'] == key]['Время замера'].min())):
            drop_keys.append(key)
            

for key in list(df_temp['key'].unique()):
    for column in df_bulk_time.columns:
        if column[0] == 'B':
            if (df_temp[df_temp['key'] == key]['Время замера'].max() < df_bulk_time[df_bulk_time['key'] == key][column].max()):
                drop_keys.append(key)
                
for key in list(df_temp['key'].unique()):
    for column in df_wire_time.columns:
        if column[0] == 'W':
            if (df_temp[df_temp['key'] == key]['Время замера'].max() < df_wire_time[df_wire_time['key'] == key][column].max()):
                drop_keys.append(key)
print(drop_keys, len(drop_keys))
```

<div style="border-radius: 15px; box-shadow: 2px 2px 2px; border: 1px solid green; padding: 20px">
We found 184 batches that do not have a final measurement temperature. These are batches with which any actions were carried out (heating, addition of bulk or wire materials) after the last temperature measurement.
</div>

```python
df_temp = df_temp.query('key not in @drop_keys')
df_temp = df_temp.dropna()
df_temp['Температура'] = pd.to_numeric(df_temp['Температура'], downcast='integer')
df_temp.info()
```

```python
# Table of results
results = pd.DataFrame({'data_type': [],
                        'model' : [], 
                        'mae_train' : [], 
                        'train_duration' : [],
                        'best_estimator' : []})
```

```python
# Linear Regressions
line = [LinearRegression(), BayesianRidge(), LassoLars()]

# Decision Tree Regressor
tree = [DecisionTreeRegressor()]
tree_params = [{'random_state': [random_state],
                'criterion':['mse'],
                'max_depth' : [x for x in range(2, 11)],
                'min_samples_split': [x for x in range(2, 6)],
                'min_samples_leaf': [x for x in range(1, 6)]}]

# Random Forest Regressor
forest = [RandomForestRegressor(n_jobs=-1)] 
forest_params = [{'criterion':['mse'],
                  'n_estimators':[x for x in range(150, 501, 25)], 
                  'random_state':[random_state]}]

# Gradient Boosting Regressor
gb = [GradientBoostingRegressor(verbose=False)]
gb_params = [{'random_state':[random_state],
              'learning_rate':[x/100 for x in range(1, 51,10)]}]


# eXtra Gradient Boosting Regressor
xgbr = [xgb.XGBRegressor(n_estimators=100,verbosity=0)]
xgbr_params = [{'learning_rate':[0.1,0.25],
                'random_state':[random_state]}]

# Cat Boost Regressor
cat = [CatBoostRegressor(verbose=False)]

# Light  Gradient Boosting Model Regressor
lgbmr = [lgb.LGBMRegressor(n_estimators=100,n_jobs=-1)]
lgbmr_params = {'learning_rate': [0.1,0.25],
                'num_leaves':[x for x in range(10,31)],
                'random_state':[random_state]}
```

```python
cv = 5
def learn_models(models_list, 
                 features_train, 
                 target_train, 
                 data_type, 
                 params={}, 
                 regr_name=''):
        # Select a model from the list, select its parameters
    for model in models_list:
        regr_gs = GridSearchCV(model, params, cv=cv, scoring='neg_mean_absolute_error')
        regr_gs.fit(features_train,target_train)
            
        cv_score = regr_gs.best_score_ *-1
        
        # Train the model, record the start time of training
        regr = regr_gs.best_estimator_
        learning_start = time.time()
        regr.fit(features_train,target_train)
        learning_end = time.time()
        learning_time = learning_end - learning_start
        
        if regr_name == '':
            name = str(model).split(sep='(')[0]
        else: name = regr_name
            
        try:    
        # Plotting a Feature Importance Analysis Graph for a Trained Model
            plt.figure(figsize=(16,5))
            ax = sns.barplot(x=features_train.columns, y=regr.feature_importances_)
            ax.set(xlabel='Feature', ylabel='Importance', title=f'Feature Importance Analysis for Model {name} Training')
            ax.set_xticklabels(features_train.columns, rotation=20, ha='right')
            ax.grid(True)
            plt.show()
        except: pass
        
        # Save the best hyperparameters

        best_params = regr_gs.best_params_
        
        # Write the results to the results table
        
        globals()['results'] = globals()['results'].append({
            'data_type' : data_type,
            'model' : name, 
            'mae_train' : round(cv_score, 5), 
            'train_duration' :  round(learning_time, 2),
            'best_estimator' : regr}, ignore_index=True)
    return plt.show()
```
